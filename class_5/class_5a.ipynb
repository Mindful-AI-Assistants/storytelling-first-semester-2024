## Trabalhando com subplots

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
```

```python
plt.figure()

plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

```python
plt.figure()

plt.subplot(1,2,1)
plt.plot(x, np.sin(x))

plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
```

```python
plt.figure()

plt.subplot(2,2,1)
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
plt.grid()

plt.subplot(2,2,3)
plt.plot(x, np.tan(x))

plt.subplot(2,2,4)
plt.plot(x, np.exp(x))
```

Colocando uma figura com resoluo especfica

```python
import matplotlib.pyplot as plt
text_kwargs = dict(ha='center', va='center', fontsize=28, color='C1')

px = 1/plt.rcParams['figure.dpi']  # pixel in inches
plt.subplots(figsize=(600*px, 200*px))
plt.text(0.5, 0.5, '600px x 200px', **text_kwargs)
plt.show()
```

## Mudando cores e estilo


```python
plt.plot(x, np.sin(x-0), color='blue') # cor por extenso
plt.plot(x, np.sin(x-1), color='g') # atalho
plt.plot(x, np.sin(x-2), color='0.75') # Escala de cinza
plt.plot(x, np.sin(x-3), color='#FFDD44') # Cdigo hexadecimal
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3))
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x-1, linestyle='-')
plt.plot(x, x-2, linestyle='--')
plt.plot(x, x-3, linestyle='-.')
plt.plot(x, x-4, linestyle=':')
```

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```

### Trabalahndo com rtulos

```python
plt.plot(x, np.sin(x))
plt.title('Curva do seno')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python

plt.plot(x, np.sin(x), '-g', label= r'sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```

## Grafico de barras

```python
data = np.random.randn(1000)
```

```python
plt.hist(data, bins=30, edgecolor='black')
plt.show()
```

```python
plt.hist(data, bins=30, color='steelblue', alpha=0.5,  edgecolor='none')
plt.show()
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=30)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]

bar_colors = ['green', 'blue', 'red', 'orange']
bar_labels = ["green", 'blue', 'red', 'orange']
plt.bar(fruits, counts, label=bar_labels, color=bar_colors)
plt.legend(title='Cor da fruta')
```

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]

bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange'] # cor da paleta
bar_labels = ["red", 'blue', '_red', 'orange'] # omitir duplicidade
plt.bar(fruits, counts, label=bar_labels, color=bar_colors)
plt.legend(title='Cor da fruta')
```
