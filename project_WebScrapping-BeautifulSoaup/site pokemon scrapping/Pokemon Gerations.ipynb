# **Importando Bibliotecas e Baixando o HTML do site**

```python
import pandas as pd
import requests
from tqdm import tqdm
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://pokemondb.net/pokedex/all' # Definindo o link do site
soup = requests.get(url) # Adquirindo o conteúdo do site
soup = BeautifulSoup(soup.content, 'html.parser') # Transformando o conteúdo em um objeto BeautifulSoup
```

# **Definindo as Características de cada Pokemon**

```python
pokedex = soup.find('tbody').find_all("tr") # Adquirindo o Conteúdo de cada pokemon
```

```python
pokedex
```

```python
# Definindo as informações do pokemon
generations = []
name = []
list_types = []
hp = []
attack = []
defense = []
sp_Atk = []
sp_Def = []
speed = []
full_power = []
img = []
```

>Acima, estamos definindo uma variável para cada informação do pokemon

```python
for pokemon in tqdm(pokedex):
    if not pokemon.find("small") is None:
        name.append(pokemon.find("small", class_ = "text-muted").text)
    
    else:
        name.append(pokemon.find("a", class_ = 'ent-name').text)
    
    aux = pokemon.find_all("a")[1:]
    qtd_power = len(aux)
    types = ''
    for power in range(qtd_power):
        types += aux[power].text
        
        if power < qtd_power - 1 and qtd_power > 1: types += ' e '
    
    list_types.append(types)
    full_power.append(int(pokemon.find("td", class_ = 'cell-num cell-total').text))
    img.append(pokemon.find("img", class_ = 'img-fixed icon-pkmn')["src"])
    
    skill = pokemon.find_all("td", class_ = 'cell-num')[2:]
    
    hp.append(int(skill[0].text))
    attack.append(int(skill[1].text))
    defense.append(int(skill[2].text))
    sp_Atk.append(int(skill[3].text))
    sp_Def.append(int(skill[4].text))
    speed.append(int(skill[5].text))
```

>Acima, estamos colocando cada informação de pokemon em sua respectiva variável

# **Criando a Base de Dados**

```python
pokedex = pd.DataFrame({"Name": name, "Type": list_types, "HP": hp, "Attack": attack, "Defense": defense, "Sp_Atk": sp_Atk,
                        "Sp.Def": sp_Def, "Speed": speed, "Total": full_power, "Photo of the Pokemon": img})
```

# **Salvando a base de dados como csv**

```python
pokedex.to_csv("Pokedex.csv", index = False)
```
